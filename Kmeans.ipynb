```python
class Kmeans:

    '''Implementing Kmeans algorithm.'''

    def __init__(self, n_clusters, epochs=100):
        self.n_clusters = n_clusters
        self.epochs = epochs
        self.centroids = np.array([])

    def _initalize_centriods(self, features):
      centriods = []
      for elem in range(self.n_clusters):
        centriods.append(np.random.rand(features.shape[1]))

      return np.array(centriods)

    def _compute_distance(self, features, centriods):
      return self._euclidean(features, centriods)
     


    def _compute_new_centroids(self, features, labels):
      unique_values, counts = np.unique(labels, return_counts=True)
      #print("Unique values : ", unique_values, "cOUNTS: ", counts)
      new_centroids = []

      for i, value in enumerate(unique_values):
        #print(i, value)
        centroid = 0
        for j, label in enumerate(labels):
         # print(j, label)
          if value == label:
            centroid += features[j] 
        new_centroids.append(centroid/ counts[i])


      return np.array(new_centroids)
        
       

    def get_closest_cluster(self, distance):
      return np.argmin(distance, axis=1)

    def train(self, features):
      self.centroids = self._initalize_centriods(features)
      self.centroids = C
      #print(".self.centroids", self.centroids)
      for i in range(self.epochs):
        old_centroids =  copy.deepcopy(self.centroids)
        c_distances = self._euclidean(features, old_centroids)
        current_labels = self.get_closest_cluster(c_distances)
        self.centroids = self._compute_new_centroids(features, current_labels)
        # print("new : ", self.centroids)
        # print("old : ", old_centroids)
        # print("equal", np.array_equal(self.centroids, old_centroids))
        if np.array_equal(self.centroids, old_centroids):
          return i, self.n_clusters, c_distances, current_labels, self.centroids 
        # else:
        #   self.centriods = new_centroids
      
        

    def _euclidean(self, features, centroids):
      #print("features: ", features)
      #print("centroids", centroids)
      distances = []
      for x in features:
        centroid_distance = []
        for c in centroids:
          distance = 0
          for i in range(len(x)):
             if isinstance(x[i], str) and isinstance(c[i], str):
                 if x[i] != c[i]:
                    distance += 1
             else:
               distance += pow((x[i] - c[i]), 2)
          centroid_distance.append(pow(distance, 1/2))
        distances.append(np.array(centroid_distance))

      return distances
    
```
